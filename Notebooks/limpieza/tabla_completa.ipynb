```python
import pandas as pd
import numpy as np
```

# Concateno todas las ligas

```python
# Importo mis tablas de datos
premier = pd.read_csv('../csv/csv_limpios/premier_lp.csv', index_col=0)
liga = pd.read_csv('../csv/csv_limpios/laliga_lp.csv', index_col=0)
serie= pd.read_csv('../csv/csv_limpios/serie_lp.csv',   index_col=0)
```

```python
#Concateno las tablas
jugadores = pd.concat([premier, liga, serie])
```

# Limpieza

```python
# Agrupo los jugadores que se repiten y hago la media de sus valores
def custom_agg(x):
    if x.dtype == "float64" or x.dtype == "int64":
        if x.name == 'Edad':
            return np.max(x)
        else:
            return np.mean(x)
    elif x.dtype == "object":
        if x.name == 'Equipo':  # Asume que quieres el último equipo
            return x.iloc[-1]  # Devuelve el último valor (última temporada)
        else:
            return x.iloc[0]
    
jugadores_tot = jugadores.groupby('Jugador').agg(custom_agg)
```
