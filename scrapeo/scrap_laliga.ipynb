```python
todos_los_jugadores = []

# Bucle para recorrer todas las pÃ¡ginas
for i in range(1, 32):
    try:
        # Extract player data
        jugadores_pagina_actual = [e.text.split('\n') for e in driver.find_elements(By.CSS_SELECTOR, 'div > table > tbody > tr')]
        todos_los_jugadores.extend(jugadores_pagina_actual)
        print(f'Pagina {i} de 31 extraida')

        # Click the next page button
        pagination_buttons = driver.find_elements(By.CSS_SELECTOR, 'div.styled__PaginationItemContainer-sc-1pb0xud-0.kqZddV')
        if i < 31:  
            pagination_buttons[i].click()  
            print('boton clickado correctamente')
            time.sleep(10)  # Adjust the sleep time if necessary
    except Exception as e:
        print(f"An error occurred on page {i}: {e}")
    
# Make sure to close the driver after the operation
driver.close()
    
```
