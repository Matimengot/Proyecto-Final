#### Limpieza


```python
# Ya se borro el primer nivel de columnas en el script de extraccion
# En mi columna Pais, me voy a quedar solo con el nombre del pais
jugadores_2023['País'] = jugadores_2023['País'].str.split(' ', expand=True)[1]
```

```python
jugadores_2023.columns
```

```python
# Elimino las columnas que no me interesan para mi estudio
jugadores_2023.drop(['RL','PrgC','PrgP','PrgR','xG.1','xAG.1','xG+xAG','npxG.1','npxG+xAG.1','Partidos','90 s','TPint','npxG','npxG+xAG','Gls..1'
                     ,'Ast','G+A.1','G-TP.1','G+A-TP','G-TP','TP'], axis=1, inplace=True)
```

```python
#Limpio la columna de edad
jugadores_2023['Edad'] = jugadores_2023['Edad'].str.split('-', expand=True)[0]
```

```python
#Veo si hay algun valor nulo
jugadores_2023.isnull().sum()
```

```python
#Veo que filas tienen pais nulo
jugadores_2023[jugadores_2023['País'].isnull()]
#Como veo que no tienen sentido los datos, los elimino
jugadores_2023.dropna(inplace=True)
#Ya no hay nulos
```

```python
jugadores_2023['Posc'].unique()
```

```python
# Filtro la columna Posc para que me queden solo cuatro grupos: Portero, Defensa, Centrocampista y Delantero
jugadores_2023['Posc'] =jugadores_2023['Posc'].replace({
    'DF': 'Defensa',
    'CC,DL': 'Centrocampista',
    'CC': 'Centrocampista',
    'DF,CC': 'Defensa',
    'DL,CC': 'Delantero',
    'DF,DL': 'Defensa',
    'DL': 'Delantero',
    'CC,DF': 'Centrocampista',
    'DL,DF': 'Delantero'})
```

```python
#Veo si tiene sentido los datos de la columna Posc
jugadores_2023['Posc'].value_counts()
```
