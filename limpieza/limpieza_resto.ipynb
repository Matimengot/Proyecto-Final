```python
import pandas as pd
import numpy as np
```

## Concadeno el resto de temporadas para limpiar todas juntas


```python
jugadores_2023 = pd.read_csv('../csv/csv_raw/jugadores_2023.csv')
jugadores_2022 = pd.read_csv('../csv/csv_raw/jugadores_2022.csv')
jugadores_2021 = pd.read_csv('../csv/csv_raw/jugadores_2021.csv')
jugadores_2020 = pd.read_csv('../csv/csv_raw/jugadores_2020.csv')
jugadores_2019 = pd.read_csv('../csv/csv_raw/jugadores_2019.csv')
jugadores_2018 = pd.read_csv('../csv/csv_raw/jugadores_2018.csv')
jugadores_2017 = pd.read_csv('../csv/csv_raw/jugadores_2017.csv')
```

```python
#Concadeno los dataframes
stats = pd.concat([jugadores_2022, jugadores_2021, jugadores_2020, jugadores_2019, jugadores_2018, jugadores_2017], ignore_index=True)
```

## Limpieza

```python
# Veo la cantidad de valores nulos
stats.isnull().sum()
#Elimino las filas con valores nulos
stats = stats.dropna()
```

```python
# Ya se borro el primer nivel de columnas en el script de extraccion
# En mi columna Pais, me voy a quedar solo con el nombre del pais
stats['País'] = stats['País'].str.split(' ', expand=True)[1]
```

```python
stats
```

```python
stats.columns
```

```python
# Elimino las columnas que no me interesan para mi estudio
stats.drop(['RL','PrgC','PrgP','PrgR','xG.1','xAG.1','xG+xAG','npxG.1','npxG+xAG.1','Partidos','90 s','TPint','npxG','npxG+xAG','Gls..1'
                     ,'Ast','G+A.1','G-TP.1','G+A-TP','G-TP','TP'], axis=1, inplace=True)
```

```python
stats
```

```python
#Limpio la columna de edad
stats['Edad'] = stats['Edad'].str.split('-', expand=True)[0]
```

```python
#Veo si hay algun valor nulo
stats.isnull().sum()
stats.dropna(inplace=True)
```

```python
stats['Posc'].unique()
```

```python
# Filtro la columna Posc para que me queden solo cuatro grupos: Portero, Defensa, Centrocampista y Delantero
stats['Posc'] =stats['Posc'].replace({
    'DF': 'Defensa',
    'CC,DL': 'Centrocampista',
    'CC': 'Centrocampista',
    'DF,CC': 'Defensa',
    'DL,CC': 'Delantero',
    'DF,DL': 'Defensa',
    'DL': 'Delantero',
    'CC,DF': 'Centrocampista',
    'DL,DF': 'Delantero'})
```

```python
stats['Posc'].value_counts()
```

```python
stats = stats.rename(columns={'Gls': 'Goles','Ass': 'Asistencias','G+A': 'Goles+Asistencias','TA': 'Tarjetas Amarillas',
                               'TR': 'Tarjetas Rojas','xG': 'Goles Esperados','xAG':'Asistencias Esperadas'})
```
