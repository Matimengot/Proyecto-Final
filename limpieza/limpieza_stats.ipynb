# Limpieza

```python
# Averiguo si hay nulos
stats.isnull().sum()
```

```python
# Veo la columna valor de mercado para ver cuando con '-' hay
stats[stats['Valor Mercado'] == '-']
```

```python
# Elimino las filas que tienen '-' en la columna valor de mercado
stats = stats[stats['Valor Mercado'] != '-']
```

```python
# Elimino la columna jugadores ya que en el modelo predictivo solo importan las estadisticas
stats = stats.drop(['Jugador'], axis=1)
```

```python
#Dropeo la columna unnamed
stats = stats.drop(['Unnamed: 0'], axis=1)
```

```python
#reseteo el index
stats.reset_index(drop=True, inplace=True)
```
