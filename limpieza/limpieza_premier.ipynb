# Limpieza

```python
# Ya se borro el primer nivel de columnas en el script de extraccion
# En mi columna Pais, me voy a quedar solo con el nombre del pais
premier['País'] = premier['País'].str.split(' ', expand=True)[1]
```

```python
premier.columns
```

```python
# Elimino las columnas que no me interesan para mi estudio
premier.drop(['RL','PrgC','PrgP','PrgR','xG.1','xAG.1','xG+xAG','npxG.1','npxG+xAG.1','Partidos','90 s','TPint','npxG','npxG+xAG','Gls..1'
                     ,'Ast','G+A.1','G-TP.1','G+A-TP','G-TP','TP'], axis=1, inplace=True)
```

```python
#Limpio la columna de edad
premier['Edad'] = premier['Edad'].str.split('-', expand=True)[0]
```

```python
#Veo si hay algun valor nulo
premier.isnull().sum()
```

```python
#Como veo que no tienen sentido los datos, los elimino
premier.dropna(inplace=True)
#Ya no hay nulos
```

```python
premier['Posc'].unique()
```

```python
# Filtro la columna Posc para que me queden solo cuatro grupos: Portero, Defensa, Centrocampista y Delantero
premier['Posc'] =premier['Posc'].replace({
    'DF': 'Defensa',
    'CC,DL': 'Centrocampista',
    'CC': 'Centrocampista',
    'DF,CC': 'Defensa',
    'DL,CC': 'Delantero',
    'DF,DL': 'Defensa',
    'DL': 'Delantero',
    'CC,DF': 'Centrocampista',
    'DL,DF': 'Delantero'})
```

```python
premier['Posc'].value_counts()
```

```python
premier = premier.rename(columns={'País':'Pais','Posc':'Posicion','Gls.': 'Goles','Ass': 'Asistencias','G+A': 'Goles+Asistencias','TA': 'Tarjetas Amarillas',
                               'TR': 'Tarjetas Rojas','xG': 'Goles Esperados','xAG':'Asistencias Esperadas'})
```

```python
premier = premier.reset_index(drop=True)
premier
```

# Exporto a csv limpio


```python
premier.to_csv('../csv/csv_limpios/premier_lp.csv')
```
