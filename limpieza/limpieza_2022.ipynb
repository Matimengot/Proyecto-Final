```python
# Ya se borro el primer nivel de columnas en el script de extraccion
# En mi columna Pais, me voy a quedar solo con el nombre del pais
jugadores_2022['País'] = jugadores_2022['País'].str.split(' ', expand=True)[1]
```

```python
jugadores_2022.columns
```

```python
# Elimino las columnas que no me interesan para mi estudio
jugadores_2022.drop(['RL','PrgC','PrgP','PrgR','xG.1','xAG.1','xG+xAG','npxG.1','npxG+xAG.1','Partidos','90 s','TPint','npxG','npxG+xAG','Gls..1'
                     ,'Ast','G+A.1','G-TP.1','G+A-TP','G-TP','TP'], axis=1, inplace=True)
```

```python
#Limpio la columna de edad
jugadores_2022['Edad'] = jugadores_2022['Edad'].str.split('-', expand=True)[0]
```

```python
#Veo si hay algun valor nulo
jugadores_2022.isnull().sum()
```

```python
#Veo que filas tienen pais nulo
jugadores_2022[jugadores_2022['País'].isnull()]
#Como veo que no tienen sentido los datos, los elimino
jugadores_2022.dropna(inplace=True)
#Ya no hay nulos
```

```python
jugadores_2022['Posc'].unique()
```

```python
# Filtro la columna Posc para que me queden solo cuatro grupos: Portero, Defensa, Centrocampista y Delantero
jugadores_2022['Posc'] =jugadores_2022['Posc'].replace({
    'DF': 'Defensa',
    'CC,DL': 'Centrocampista',
    'CC': 'Centrocampista',
    'DF,CC': 'Defensa',
    'DL,CC': 'Delantero',
    'DF,DL': 'Defensa',
    'DL': 'Delantero',
    'CC,DF': 'Centrocampista',
    'DL,DF': 'Delantero'})
```

```python
jugadores_2022['Posc'].value_counts()
```

```python
jugadores_2022 = jugadores_2022.rename(columns={'Gls': 'Goles','Ass': 'Asistencias','G+A': 'Goles+Asistencias','TA': 'Tarjetas Amarillas',
                               'TR': 'Tarjetas Rojas','xG': 'Goles Esperados','xAG':'Asistencias Esperadas'})
```

```python
#Borro los jugadores que estan dos veces
jugadores_2022 = jugadores_2022.drop_duplicates(subset='Jugador', keep='first').reset_index(drop=True)
```
