# Limpieza

```python
# Ya se borro el primer nivel de columnas en el script de extraccion
# En mi columna Pais, me voy a quedar solo con el nombre del pais
liga['País'] = liga['País'].str.split(' ', expand=True)[1]
```

```python
liga.columns
```

```python
# Elimino las columnas que no me interesan para mi estudio
liga.drop(['RL','PrgC','PrgP','PrgR','xG.1','xAG.1','xG+xAG','npxG.1','npxG+xAG.1','Partidos','90 s','TPint','npxG','npxG+xAG','Gls..1'
                     ,'Ast','G+A.1','G-TP.1','G+A-TP','G-TP','TP'], axis=1, inplace=True)
```

```python
#Limpio la columna de edad
liga['Edad'] = liga['Edad'].str.split('-', expand=True)[0]
```

```python
#Veo si hay algun valor nulo
liga.isnull().sum()
```

```python
#Como veo que no tienen sentido los datos, los elimino
liga.dropna(inplace=True)
#Ya no hay nulos
```

```python
liga['Posc'].unique()
```

```python
# Filtro la columna Posc para que me queden solo cuatro grupos: Portero, Defensa, Centrocampista y Delantero
liga['Posc'] =liga['Posc'].replace({
    'DF': 'Defensa',
    'CC,DL': 'Centrocampista',
    'CC': 'Centrocampista',
    'DF,CC': 'Defensa',
    'DL,CC': 'Delantero',
    'DF,DL': 'Defensa',
    'DL': 'Delantero',
    'CC,DF': 'Centrocampista',
    'DL,DF': 'Delantero'})
```

```python
liga['Posc'].value_counts()
```

```python
liga = liga.rename(columns={'País':'Pais','Posc':'Posicion','Gls.': 'Goles','Ass': 'Asistencias','G+A': 'Goles+Asistencias','TA': 'Tarjetas Amarillas',
                               'TR': 'Tarjetas Rojas','xG': 'Goles Esperados','xAG':'Asistencias Esperadas'})
```

```python
liga = liga.reset_index(drop=True)
liga
```
