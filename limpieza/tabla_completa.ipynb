```python
import pandas as pd
import numpy as np

```

# Concateno todas las ligas

```python
# Importo mis tablas de datos
premier = pd.read_csv('../csv/csv_limpios/premier_lp.csv', index_col=0)
liga = pd.read_csv('../csv/csv_limpios/laliga_lp.csv', index_col=0)
serie= pd.read_csv('../csv/csv_limpios/serie_lp.csv',   index_col=0)
```

```python
#Concateno las tablas
jugadores = pd.concat([premier, liga, serie])
```

# Limpieza

```python
# Agrupo los jugadores que se repiten y hago la media de sus valores
def custom_agg(x):
    if x.dtype == "float64" or x.dtype == "int64":
        if x.name == 'Edad':
            return np.max(x)
        else:
            return np.mean(x)
    elif x.dtype == "object":  # Asume que el tipo de objeto es una cadena
        return x.iloc[0]
    
jugadores_tot = jugadores.groupby('Jugador').agg(custom_agg)
```

```python
jugadores_tot['Valor de mercado'].value_counts()
```

```python
# Saco los valores '-' de la columna valor
jugadores_tot = jugadores_tot[jugadores_tot['Valor de mercado'] != '-']
```

```python
jugadores_tot
```

```python
# Paso los valores de la columna valor a float
def valor_mercado(value):

    value = value.replace('€', '').replace(',', '.').strip()
    parts = value.split()
    
    if len(parts) == 2:
        number, unit = parts
        if unit == 'mill.':
            return float(number) * 1e6
        elif unit == 'mil':
            return float(number) * 1e3
    else:
        return float(value)
```

```python
jugadores_tot['Valor de mercado'] = jugadores_tot['Valor de mercado'].apply(valor_mercado)
```

```python
#Reenombro la columna posicion
jugadores_tot.rename(columns={'Posc': 'Posicion'}, inplace=True)
```
