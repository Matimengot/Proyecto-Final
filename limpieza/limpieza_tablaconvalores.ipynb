# Limpieza

```python
# Estudio la columna valores de mercado
tabla['Valor de mercado'].unique()
```

```python
# Elimino los valores de mercado que sean '-'
tabla = tabla[tabla['Valor de mercado'] != '-']
```

```python
def valor_mercado(value):

    value = value.replace('€', '').replace(',', '.').strip()
    parts = value.split()
    
    if len(parts) == 2:
        number, unit = parts
        if unit == 'mill.':
            return float(number) * 1e6
        elif unit == 'mil':
            return float(number) * 1e3
    else:
        return float(value)
```

```python
tabla['Valor de mercado'] = tabla['Valor de mercado'].apply(valor_mercado)
```

```python
tabla['Valor de mercado'].unique()
```

```python
# Veo si hay algun valor de mercado igual a 0
tabla[tabla['Valor de mercado'] == 0.0]
```

```python
# Elimino los valores de mercado que sean 0
tabla = tabla[tabla['Valor de mercado'] != 0.0]
```
